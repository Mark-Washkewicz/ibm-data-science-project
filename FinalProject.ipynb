<h2> Data </h2>

<p>The data gathered for this project was the different neighborhoods and postcodes in Toronto (Wikipedia), location data on these neighborhoods, <a href="https://www12.statcan.gc.ca/census-recensement/2016/dp-pd/hlt-fst/pd-pl/Table.cfm?Lang=Eng&T=1201&SR=1&S=22&O=A&RPP=9999&PR=0"> census data </a> on the population in these different neighborhoods, and the number of coffee shops in these neighborhoods.</p>

<p> Once the data was gathered the dataframes had to be cleaned and conglomerated so that the number of coffee shops per capita could be found. This was done by combining the CSV from the Canadian government with Foursquare and Location data. It should be noted that certain neighborhoods were removed due to their low populations that would skew the per capita, and as we only care about other popular coffee shops, we only took the coffee shops that were in the top 500 squarespace venues. Once all of this data is combined, the coffee shops per capita for each neighborhood can be calculated and the lowest coffee shops per capita is where a coffee shop should open.</p>

<h2> Methodology </h2>

<p> The different data was read using different read functions from pandas. The postal codes and neighborhoods were read using the read_html funciton and the csv for neighborhoods was read_csv. By using these two functions, coupled with merge. We are able to combine the data into one new df. Which allows us to see the population for each neighborhood along with the number of popular coffee shops. We then take the per capita number of coffee shops by using the pandas div function. And then finally we sort by the per capita in ascending order to determine which area has the least number of coffee shops per capita </p>

```python
def getNearbyVenues(names, latitudes, longitudes, search_query, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        
        
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```
